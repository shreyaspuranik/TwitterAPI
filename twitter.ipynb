```python
#Function to retrieve most recent tweets of a user
def all_tweets(screen_name):
    import time
    list1 = []
    user_timeline = twitter.get_user_timeline(screen_name= screen_name,count = 1)
    lis = [user_timeline[0]["id"]]## this is the latest starting tweet id
    for i in range(0,17): ## iterate through all tweets
    ## tweet extract method with the last list item as the max_id
        user_timeline = twitter.get_user_timeline(screen_name= screen_name,
        count=200, include_retweets=False, max_id=lis[-1])
        #time.sleep(60)

        for tweet in user_timeline:
            list1.append(tweet['text'])
            lis.append(tweet['id']) ## append tweet id's
    
    list1 = list(unique_everseen(list1))
    return list1

list_tweets = all_tweets("BJP4India")
```

```python
#function to retrieve particular features from a tweet
def tweet_features(tweet):
    created_at = tweet.get("created_at") #date and time at which tweet was created
    
    ts = time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(tweet.get("created_at"),'%a %b %d %H:%M:%S +0000 %Y'))
    created_at_time = ts[11:]
    created_at_date = ts[:10]
    created_at_time_off = created_at[20:25]
    hashtags = ",".join([tweet.get('entities').get('hashtags')[i].get('text') for i in range(len(tweet.get('entities').get('hashtags')))])
    links = ",".join(re.findall(r'(https?://[^\s]+)',tweet["text"] ))
    mentions = ",".join(re.findall(r'(@[^\s]+)',tweet["text"] ))
    text = tweet["text"]
    created_at_day = created_at[0:3]
    created_at_month = created_at[4:7]
    favourites_count = tweet.get("user").get("favourites_count")
    lang = tweet.get("lang")
    retweet_count = tweet.get("retweet_count")
    source = tweet.get("source")
    
    return [created_at, created_at_time, created_at_date,  created_at_time_off, hashtags, links, mentions, text, \
            created_at_day, created_at_month, favourites_count, lang, retweet_count, source]
```
